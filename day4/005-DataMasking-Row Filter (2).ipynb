```sql
-- this example to be used in sql warehouse, but can be used in serverless/cluster notebooks too
-- copy paste as instructor advise

CREATE DATABASE IF NOT EXISTS  usersdb;

CREATE OR REPLACE TABLE  usersdb.user_country_data (
    id INT,
    name STRING,
    email STRING,
    country STRING
);

INSERT INTO usersdb.user_country_data VALUES
  (1, 'Alice', 'alice@example.com', 'USA'),
  (2, 'Ravi', 'ravi@example.com', 'India'),
  (3, 'John', 'john@example.com', 'UK'),
  (4, 'Priya', 'priya@example.com', 'India');

-- short circuit OR 
-- return value is boolean, if true means the row is allowed to be seen by user/developer
-- return value is false, user cannot see the row 
-- if user is primary user, who can see all teh data
-- if user is secondary user, who can see only India data
CREATE OR REPLACE FUNCTION usersdb.row_filter_country(country STRING)
RETURNS BOOLEAN
RETURN
  CURRENT_USER() = 'gs@training.sh' OR 
  (CURRENT_USER() = 'student2@gopalakrishnan.onmicrosoft.com' AND country = 'India');

SELECT * FROM usersdb.user_country_data;



GRANT SELECT ON TABLE usersdb.user_country_data TO `student2@gopalakrishnan.onmicrosoft.com`;

-- on second user, who can see all data
-- NOTE, row filter not yet applied
SELECT * FROM usersdb.user_country_data;

ALTER TABLE usersdb.user_country_data 
SET ROW FILTER   usersdb.row_filter_country ON (country);

-- now again on second user, after we applied row filter

SELECT * FROM usersdb.user_country_data;



 
```
