```sql
-- to be used in sql query editor
-- you need to change path as per instructor instructions

-- https://docs.databricks.com/aws/en/data-governance/unity-catalog/manage-privileges/privileges
SHOW CREDENTIALS;

SHOW EXTERNAL LOCATIONS;

-- try accessing one of the storage location data in anotehr user before grant permission

SELECT * FROM csv.`abfss://bronze@gksdatalake2.dfs.core.windows.net/movies`;

DESCRIBE EXTERNAL LOCATION `gks2`;

GRANT ALL PRIVILEGES
ON EXTERNAL LOCATION gks2
TO `student2@gopalakrishnan.onmicrosoft.com`;

GRANT USE SCHEMA 
ON SCHEMA  deloitte_db.gks_movielens
TO `student2@gopalakrishnan.onmicrosoft.com`;

GRANT SELECT
ON TABLE deloitte_db.gks_movielens.bronze_movies
TO `student2@gopalakrishnan.onmicrosoft.com`;

SHOW GRANTS ON TABLE deloitte_db.gks_movielens.bronze_movies;

-- run below queries on studen2/another user

SHOW EXTERNAL LOCATIONS;
-- permission denied to user
SELECT * FROM csv.`abfss://bronze@gksdatalake2.dfs.core.windows.net/movies`;



REVOKE ALL PRIVILEGES
ON EXTERNAL LOCATION gks2
FROM `student2@gopalakrishnan.onmicrosoft.com`;


-- on other users, 
SHOW EXTERNAL LOCATIONS;

-- however still user can access table, since he/she granted with select permision on the table

-- permission on external location helps a user to create a external tables/write data to external location

```
